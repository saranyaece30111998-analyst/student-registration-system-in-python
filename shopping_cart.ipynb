Case Study – Shopping Cart System
Instructions:

Create a list named cart with initial items: Milk, Bread, Eggs, Apple

Perform list operations:

• Add multiple items using extend()

• Insert item at index 2 using insert()

• Remove item Apple using remove()

• Remove last item using pop()

Use list methods:

• index() to find position of Eggs

• count() to count occurrences of Milk

Apply indexing:

• Display first item

• Display last item

Apply slicing:

• Display first 3 items from the list

Find total number of items using len()

Display all outputs using print()


```python
# creating card
cart = ["Milk", "Bread", "Eggs", "Apple"]
print("Initial cart:", cart)

# Add multiple items
cart.extend(["Butter", "Cheese"])
print("After extend:", cart)

# Insert item
cart.insert(2, "Juice")
print("After insert at index 2:", cart)

# Remove item Apple
cart.remove("Apple")
print("After removing Apple:", cart)

# Remove last item
cart.pop()
print("After removing last item:", cart)

#   find position of Eggs
print("Index of Eggs:", cart.index("Eggs"))

# count occurrences of Milk
print("Count of Milk:", cart.count("Milk"))

# Apply indexing
print("First item:", cart[0])
print("Last item:", cart[-1])

# Apply slicing
print("First 3 items:", cart[:3])

# Find total number ()
print("Total items:", len(cart))
```
