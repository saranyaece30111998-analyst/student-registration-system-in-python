## **Python Assignment 1 - Data Structures - Strings & Tuples **


```python
#1.String Concatenation:
# concatenation of two string

string_1="hello"
string_2=input("enter your name:")
new_string=string_1+" "+string_2
print(f"new string is:{new_string}")

#concatenat one more string
string_3="welcome to python programming"
final_string=new_string+", "+string_3
print(f"final string is:{final_string}")


  #2. String Slicing and Indexing:
text="hello saranya welcome to python programming"

#  Print the first character of the string
print(f"First character: {text[0]}")

#  Print the last character of the string
print(f"Last character: {text[-1]}")

#  Print the first 5 characters of the string
print(f"First 5 characters:{ text[:5]}")

#  Print the last 11 characters of the string
print(f"Last 11 characters:{ text[-11:]}")

#  Print the string in reverse
print(f"Reversed string:{ text[::-1]}")

#  Use slicing and print the word “Python” from the existing string
print(f"Extracted word: {text[25:31]}")


```

```python
# 3.string method
#Input string
strM = "Python beginner tutorial"

# a. Convert the sentence to uppercase
print(f"Uppercase:{ strM.upper()}")

# b. Convert the sentence to lowercase
print(f"Lowercase:{ strM.lower()}")

# c. Use Capitalize and return the sentence to the original input form
print(f"Capitalize:{ strM.capitalize()}")

# d. Count the total number of occurrences of character ‘t’
print(f"Occurrences of 't':{ strM.count('t')}")

# e. Replace all occurrences of “Python” with “Data Analytics”
print(f"Replaced string: {strM.replace("Python", "Data Analytics")}")
```

```python
# Tuples (Creation, Modification and Access) :
#Create tuples
tuple1 = (10, 20, 30)
tuple2 = (40, 50, 60)

#  Concatenate the two tuples
t_combine = tuple1 + tuple2
print(f"Concatenated tuple:{ t_combine}")

#  Repeat the elements of “t_combine” 3 times
print(f"Repeated tuple:{t_combine * 3}")

# c. Access the 3rd element from “t_combine”
print(f"3rd element:{ t_combine[2]}")

# d. Access the first three elements from “t_combine”
print(f"First three elements:{ t_combine[:3]}")

# e. Access the last three elements from “t_combine”
print(f"Last three elements:{ t_combine[-3:]}")
```
