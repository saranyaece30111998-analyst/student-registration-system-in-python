Case Study – Text Analyzer System
Instructions:

Take a string input from the user using input().

Convert the text into lowercase using lower().

Use a for loop to iterate through each character in the string.

Count: • Total vowels
• Total consonants
• Total digits
• Total special characters

Use: • isalpha() for alphabets
• isdigit() for numbers

Display all counts using print()

```python
# text analyzer system


text = input("enter your text: ").lower()

vowels = 0
consonants = 0
digits = 0
special_chars = 0


for ch in text:
    if ch in "aeiou":
        vowels += 1
    elif ch.isalpha():
        consonants += 1
    elif ch.isdigit():
        digits += 1
    else:
        special_chars += 1


print(f"total vowels:{vowels}")
print(f"total consonants:{ consonants}")
print(f"total digits:{ digits}")
print(f"total special characters:{ special_chars}")
```

```python
# Case Study – Restricted Word Detection System

# List of words
words = ["Hello", "Python", "Admin", "Blocked", "Welcome"]

# Restricted word
restricted_word = "Blocked"

# for loop
for word in words:

    # Checking restricted word
    if word == restricted_word:
        print(f"Restricted word detected found: {word}")
        break

    # Safe words
    print(f"Safe words:{word}")# Case Study – Restricted Word Detection System
```

```python

words = ["hello", "python", "admin", "blocked", "welcome","blocked","hacker","wrong"]

restricted = ["blocked", "hacker", "wrong"]

count = 0

for word in words:
    if word in restricted:
      print(f"restricted word found:{ word}")
      count += 1


print(f"total restricted words:{count}")
```

```python
students = ["saranya", "priya", "chitra", "divya", "eshwar"]
marks = [85, 78, 92, 67, 74]

print("using range() with index positions ")
for i in range(len(students)):
    print(f"student:{ students[i]}marks:{ marks[i]}")
```

```python
print("using enumerate() to show index and name ")
for index, name in enumerate(students):
    print(f"index:{ index}student: {name}")
```

```python

for name, mark in zip(students, marks):
    print(f"student:{ name}  marks:{ mark}")
```
