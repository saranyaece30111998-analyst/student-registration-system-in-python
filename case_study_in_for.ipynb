Case Study – String Analysis using List Comprehension **Instructions **

Create a string variable called text

Store any sentence or word inside the variable

Use list comprehension to:

Convert all characters into uppercase

Extract only vowels from the string

Extract only consonants from the string

Display:

Original string

Uppercase characters

Vowels list

Consonants list

String without spaces

Print the output in a clean format using print() statements

```python
# Case Study – String Analysis using List Comprehension

# 1. Create a string variable
text = "data analysis is powerful"

# 2. Convert all characters into uppercase
uppercase_chars = [char.upper() for char in text]

# 3. Extract only vowels from the string
vowels = [char for char in text if char.lower() in 'aeiou']

# 4. Extract only consonants from the string
consonants = [char for char in text if char.isalpha() and char.lower() not in 'aeiou']

# 5. String without spaces
string_no_spaces = ''.join([char for char in text if char != ' '])

# Display the results
print(f"Original String: { text}")
print(f"Uppercase Characters: {uppercase_chars}")
print(f"Vowels List: {vowels}")
print(f"Consonants List: { consonants}")
print(f"String without spaces: {string_no_spaces}")
```

Case Study – Employee Salary Increment using List Comprehension Instructions

Create a list called salary

Store employee salaries inside the list

Use list comprehension to: Increase each salary by 10%

Create another list to: Display only salaries greater than 50000

Print:

Original salary list

Updated salary list

Filtered salary list

```python
# Case Study – Employee Salary Increment using List Comprehension

# 1. Create a list called salary
salary = [35000, 48000, 52000, 60000, 45000, 75000]

# 2. Increase each salary by 10% using list comprehension
updated_salary = [s * 1.10 for s in salary]

# 3. Create another list to display only salaries greater than 50000
filtered_salary = [s for s in updated_salary if s > 50000]

# 4. Print the results
print(f"Original Salary List: { salary}")
print(f"Updated Salary List (10% increment): { updated_salary}")
print(f"Filtered Salary List (> 50000): {filtered_salary}")
```
