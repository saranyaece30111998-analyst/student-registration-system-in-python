```python
# Tuple Packing
student = ("Saranya", 27, "Data Analyst")

print("Packed Tuple:", student)

# Tuple Unpacking
name, age, course = student

print("Name:", name)
print("Age:", age)
print("course:", course)


# CREATE A NEW TUPLe

students=(name,age,Course)
print(students)
print(type(students))
```

```python
# Create a tuple with order details
order = (101, "Laptop", 2, 55000, "Shipped")

# Unpack the tuple into separate variables
order_id, product, quantity, price, status = order

# Display all order details using print() and f-string
print(f"order ID: {order_id}")
print(f"Product Name: {product}")
print(f"quantity: {quantity}")
print(f"price: {price}")
print(f"status: {status}")
```
