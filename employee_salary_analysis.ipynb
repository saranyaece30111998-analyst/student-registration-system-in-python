Case Study – Employee Salary Analysis Instructions

Create a list called salary Store employee salaries inside the list

Use Built-in Functions to:

Find maximum salary

Find minimum salary

Find total salary

Find number of employees using len()

Use lambda with map() to: Increase each salary by 10%

Use lambda with filter() to: Display salaries greater than 50000

Display:

Original salary list

Updated salary list

High salary list

Maximum salary

Minimum salary

Total salary

```python
# Case Study – Employee Salary Analysis

#  Create a list of salaries
salary = [45000, 52000, 61000, 48000, 75000, 30000]

#  Use Built-in Functions
max_salary = max(salary)
min_salary = min(salary)
total_salary = sum(salary)
num_employees = len(salary)

#Use lambda with map() to increase each salary by 10%
updated_salary = list(map(lambda s: round(s * 1.10), salary))

#  Use lambda with filter() to display salaries greater than 50000
high_salary = list(filter(lambda s: s > 50000, updated_salary))

#  Display results
print(f"original Salary List:{salary}")
print(f"updated Salary List (10% increase):{ updated_salary}")
print(f"high Salary List (>50000):{high_salary}")
print(f"maximum Salary:{max_salary}")
print(f"minimum Salary:{ min_salary}")
print(f"total Salary:{ total_salary}")

```
